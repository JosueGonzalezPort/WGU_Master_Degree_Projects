```python
# Drop unnecessary columns
columns_to_drop = ['Customer_id', 'zip_code']  
cleaned_data = df.drop(columns=columns_to_drop, errors='ignore')
print(df)
```

```python
# Check for missing values
missing_values = cleaned_data.isnull().sum()
print(missing_values)

# Fill missing values
for column in cleaned_data.columns:
    if cleaned_data[column].isnull().sum() > 0:
        if cleaned_data[column].dtype in ['float64', 'int64']:
            cleaned_data[column].fillna(cleaned_data[column].median(), inplace=True)
        else:
            cleaned_data[column].fillna(cleaned_data[column].mode()[0], inplace=True)
```

```python
# Calculate z-scores and identify outliers
z_scores = np.abs(stats.zscore(cleaned_data.select_dtypes(include=[np.number])))
outliers = (z_scores > 3).any(axis=1)

# Separate the outliers
identified_outliers = cleaned_data[outliers]

# Save the outliers to a separate CSV file
identified_outliers.to_csv('identified_outliers.csv', index=False)

# Remove outliers from the original DataFrame
cleaned_data_no_outliers = cleaned_data[~outliers]
```
