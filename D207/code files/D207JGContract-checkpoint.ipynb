```python
#imports
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data set into Pandas dataframe
df = pd.read_csv('C:\\Users\\josue\\Desktop\\WGU\\D207\\churn_cleanedtest.csv')

print(df)
#describe the dataset
df.describe()
```

```python
#remove unwanted columns
unwanted_columns = ["Unnamed", "Interaction", "Lat", "Lng"]

for uc in unwanted_columns:
    if uc in df.columns:
        df.drop(columns=uc, inplace=True)
print (df)   
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Churn'], df['Contract'])
contingency_table 
```

```python
#get the % of out test variables
contingency_pct = pd.crosstab(df['Churn'], df['Contract'], normalize='index')
contingency_pct
```

```python
# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Threshold for significance
alpha = 0.05

# Check if the p-value is less than the significance level
is_significant = p < alpha

#Print results
print(f"Chi-Square statistic: {chi2:.4f}")
print(f"p-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)
print(f"Is the result significant at alpha={alpha}? {'Yes' if is_significant else 'No'}")
```

```python
from sklearn.preprocessing import LabelEncoder
# Encode categorical variables
label_encoder = LabelEncoder()
df['Churn_Encoded'] = label_encoder.fit_transform(df['Churn'])
df['Internet_Encoded'] = label_encoder.fit_transform(df['InternetService'])

# Create dataframe for heatmap bivariate analysis of correlation
bivariate_Statics = df[['MonthlyCharge', 'Bandwidth_GB_Year', 'Outage_sec_perweek', 'Churn_Encoded', 'Internet_Encoded']]

# Calculate the correlation matrix
correlation_matrix = bivariate_Statics.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Continuous and Encoded Categorical Variables')
plt.show()
```
